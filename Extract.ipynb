```python
'''แบบที่ 1 base'''

import fitz  # PyMuPDF
import os

pdf_path = "Sample1pdf.pdf" # ชื่อไฟล์ที่จะอ่าน น่าจะทำเป็น list กำหนดจำนวน
output_folder = "pymupdf_pdf_to_images" # โฟลเดอร์ปลายทางที่เก็บ Images

os.makedirs(output_folder, exist_ok=True)
generated_images = [] # เก็บ path Images ที่สร้าง

# เปิดอ่านไฟล์และบันทึกเป็นรูปภาพ
doc = fitz.open(pdf_path)
print(f"เปิดไฟล์ PDF สำเร็จ (มีทั้งหมด {len(doc)} หน้า)")

for page_num in range(len(doc)):
    page = doc.load_page(page_num)
    pix = page.get_pixmap(matrix=fitz.Matrix(300 / 72, 300 / 72)) # ปรับความชัด DPI = 300
    output_image_path = os.path.join(output_folder, f"page_{page_num + 1}.png")
    pix.save(output_image_path)
    generated_images.append(output_image_path)

doc.close()
print(f"แปลง PDF เป็นรูปภาพเรียบร้อยแล้ว: {generated_images}")
```

```python
'''ลองเขียนให้รับหลายไฟล์'''

import fitz  # PyMuPDF
import os

# กำหนด list ของไฟล์ PDF ที่ต้องการแปลง
pdf_files = ["Sample1pdf.pdf"]

# โฟลเดอร์ปลายทางสำหรับเก็บภาพ
output_folder = "pymupdf_pdf_to_images"
os.makedirs(output_folder, exist_ok=True)

all_generated_images = {}  # เก็บผลลัพธ์เป็น dict {pdf_file: [images]}

for pdf_path in pdf_files:
    generated_images = []  # เก็บ path ของภาพจากไฟล์นี้
    try:
        doc = fitz.open(pdf_path)
        print(f"เปิดไฟล์ {pdf_path} สำเร็จ (มีทั้งหมด {len(doc)} หน้า)")

        for page_num in range(len(doc)):
            page = doc.load_page(page_num)
            pix = page.get_pixmap(matrix=fitz.Matrix(300 / 72, 300 / 72))  # DPI = 300
            output_image_path = os.path.join(
                output_folder, f"{os.path.splitext(os.path.basename(pdf_path))[0]}_page_{page_num + 1}.png"
            )
            pix.save(output_image_path)
            generated_images.append(output_image_path)

        doc.close()
        print(f"แปลง {pdf_path} เป็นรูปภาพเรียบร้อยแล้ว: {generated_images}")
        all_generated_images[pdf_path] = generated_images
    except Exception as e:
        print(f"Error processing {pdf_path}: {e}")
```

```python
import os
import fitz  # PyMuPDF
from typhoon_ocr import ocr_document

# ==========================================
# 1. ตั้งค่าไฟล์ และเปลี่ยนไปใช้ Ollama Server
# ==========================================
pdf_name = "Sample1pdf.pdf"
output_folder = "pymupdf_pdf_to_images"
markdown_output_filename = "typhoonOCR_Images_to_markdown.md"

# 🔄 ปรับปลายทางเชื่อมต่อไปที่ Ollama (Port 11434)
# (บน Google Colab หากเชื่อมต่อเครื่อง Local ให้เปลี่ยน localhost เป็น IP เครื่องคอมฯ คุณแทน)
OCR_BASE_URL = "http://localhost:11434/v1"
OCR_API_KEY = "ollama"  # ใส่คำว่า ollama ได้เลยเมื่อใช้คู่กับ Ollama
OCR_MODEL_NAME = "scb10x/typhoon-ocr1.5-3b" # ชื่อโมเดลที่โหลดไว้ใน Ollama

# สร้างโฟลเดอร์สำหรับเก็บรูปภาพชั่วคราว
os.makedirs(output_folder, exist_ok=True)


# ==========================================
# 2. ขั้นตอนการแปลง PDF เป็นรูปภาพ (PNG)
# ==========================================
generated_images = []

print("=== เริ่มต้นกระบวนการแปลง PDF เป็นรูปภาพ ===")
try:
    doc = fitz.open(pdf_name)
    total_pages = len(doc)
    print(f"เปิดไฟล์ PDF สำเร็จ (มีทั้งหมด {total_pages} หน้า)")

    for page_num in range(total_pages):
        page = doc.load_page(page_num)
        # ปรับความชัด DPI = 300
        pix = page.get_pixmap(matrix=fitz.Matrix(300 / 72, 300 / 72))

        output_image_path = os.path.join(output_folder, f"page_{page_num + 1}.png")
        pix.save(output_image_path)
        generated_images.append(output_image_path)
        print(f"-> แปลงหน้า {page_num + 1}/{total_pages} สำเร็จ")

    doc.close()
    print("แปลงไฟล์ PDF เป็นรูปภาพทั้งหมดเรียบร้อยแล้ว!\n")

except FileNotFoundError:
    print(f"❌ ไม่พบไฟล์ '{pdf_name}' กรุณาอัปโหลดไฟล์หรือเช็กชื่อไฟล์อีกครั้ง")
    generated_images = []
except Exception as e:
    print(f"❌ เกิดข้อผิดพลาดในการแปลง PDF: {e}")
    generated_images = []


# ==========================================
# 3. ขั้นตอนการทำ OCR ด้วย Typhoon ผ่าน Ollama
# ==========================================
if generated_images:
    print(f"=== เริ่มต้นกระบวนการ OCR ด้วย Ollama ({OCR_MODEL_NAME}) ===")
    all_markdown = []

    for idx, img_path in enumerate(generated_images, start=1):
        print(f"กำลังส่ง OCR หน้าที่ {idx}/{len(generated_images)} -> {OCR_BASE_URL}")

        try:
            markdown_content = ocr_document(
                img_path,
                model=OCR_MODEL_NAME,
                figure_language="Thai",
                task_type="v1.5",
                base_url=OCR_BASE_URL,
                api_key=OCR_API_KEY
            )

            all_markdown.append(f"# Page {idx}\n{markdown_content}")
            print(f"✅ OCR หน้าที่ {idx} สำเร็จ")

        except Exception as e:
            print(f"❌ เกิดข้อผิดพลาดในหน้า {idx}: {e}")
            all_markdown.append(f"# Page {idx}\n\n*(การทำ OCR ในหน้านี้เกิดข้อผิดพลาด)*\n")

    # รวมเนื้อหาและบันทึกไฟล์
    final_markdown = "\n\n".join(all_markdown)
    with open(markdown_output_filename, "w", encoding="utf-8") as f:
        f.write(final_markdown)

    print("\n==========================================")
    print(f"🎉 OCR เสร็จสิ้นสมบูรณ์ผ่าน Ollama!")
    print(f"ผลลัพธ์ถูกบันทึกไว้ที่ไฟล์: {markdown_output_filename}")
    print("==========================================")
else:
    print("\n❌ ไม่สามารถทำ OCR ได้ เนื่องจากขั้นตอนแปลงไฟล์ PDF ไม่สำเร็จ")
```
